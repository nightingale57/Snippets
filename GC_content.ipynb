```python
def GC_content(sequence):
    '''Calculates GC content of a sequence'''
    seq = sequence
    length = len(sequence)
    g_count = sequence.count("G")
    c_count = sequence.count("C")

    gc_content = (g_count + c_count)/length
    return gc_content
```
